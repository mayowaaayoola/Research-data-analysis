```python
# Probability Distributions

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

def plot_distribution(dist_name, params, size=1000):
    if dist_name == 'normal':
        mean, std = params
        data = np.random.normal(mean, std, size)
        pdf_x = np.linspace(mean - 4*std, mean + 4*std, 100)
        pdf_y = stats.norm.pdf(pdf_x, mean, std)
    elif dist_name == 'binomial':
        n, p = params
        data = np.random.binomial(n, p, size)
        pdf_x = np.arange(0, n + 1)
        pdf_y = stats.binom.pmf(pdf_x, n, p)
    elif dist_name == 'poisson':
        lam = params[0]
        data = np.random.poisson(lam, size)
        pdf_x = np.arange(0, np.max(data) + 1)
        pdf_y = stats.poisson.pmf(pdf_x, lam)
    else:
        raise ValueError("Unsupported distribution")

    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram of data')
    plt.plot(pdf_x, pdf_y, 'r-', lw=3, label=f'{dist_name.capitalize()} PDF')
    plt.title(f'{dist_name.capitalize()} Distribution')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

# Parameters for different distributions
normal_params = (0, 1)       # mean, std
binomial_params = (10, 0.5)  # n, p
poisson_params = (5,)        # lambda

# Plot distributions
plot_distribution('normal', normal_params)
plot_distribution('binomial', binomial_params)
plot_distribution('poisson', poisson_params)
```
